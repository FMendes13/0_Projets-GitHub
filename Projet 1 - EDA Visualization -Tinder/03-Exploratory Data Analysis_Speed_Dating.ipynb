Data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

```python
tinder_profils = tinder_df[['iid', 'gender', 'idg', 'age', 
                            'field', 'field_cd', 'career', 'career_c',
                            'undergra', 'mn_sat', 'tuition', 'race', 
                            'imprace', 'imprelig', 'from', 'zipcode', 
                            'income', 'goal', 'date', 'go_out', 
                            'sports', 'tvsports', 'exercise', 'dining',
                            'museums', 'art', 'hiking', 'gaming', 'clubbing', 
                            'reading', 'tv', 'theater', 'movies', 'concerts', 
                            'music', 'shopping', 'yoga', 'exphappy', 'expnum']]

tinder_profils_activities=tinder_df[['sports', 'tvsports', 'exercise', 'dining',
                            'museums', 'art', 'hiking', 'gaming', 'clubbing', 
                            'reading', 'tv', 'theater', 'movies', 'concerts', 
                            'music', 'shopping', 'yoga', 'exphappy', 'expnum']]
```

```python
tinder_profils_gender=tinder_profils['gender'].value_counts().reset_index()
tinder_profils_gender['gender'] = tinder_profils_gender['gender'].replace({1: 'man', 0: 'woman'})

fig = px.pie(tinder_profils_gender,values='count',names='gender',
             title='Distribution des genres - Tinder Experience',
             width=400)

fig.show()
```

```python
fig = make_subplots(rows=6,cols=1)
fig.add_trace(
    go.Histogram(
        x = tinder_profils['age'],
        name = 'age'),
    row=1,
    col=1)

fig.add_trace(
    go.Histogram(
        x = tinder_profils['race'],
        name = 'race'),
    row=2,
    col=1)

fig.add_trace(
    go.Histogram(
        x = tinder_profils['imprace'],
        name = 'imprace'),
    row=3,
    col=1)

fig.add_trace(
    go.Histogram(
        x = tinder_profils['goal'],
        name = 'goal'),
    row=4,
    col=1)

fig.add_trace(
    go.Histogram(
        x = tinder_profils['date'],
        name = 'date'),
    row=5,
    col=1)

fig.add_trace(
    go.Histogram(
        x = tinder_profils['expnum'],
        name = 'expnum'),
    row=6,
    col=1)

fig.update_layout(width=700,height=1000)
fig.show()
```

```python
fig = px.box(tinder_profils, x='income', title='Distribution des revenus', 
                   width=700, height=400)
fig.show()
```
